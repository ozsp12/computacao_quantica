### Explicação teórica do circuito quântico

O circuito apresentado é composto por dois qubits (`q0` e `q1`) e uma sequência de portas que ilustram operações básicas de preparação de estado, criação de superposição, entrelaçamento e rotação. A sequência completa é:

1. **X(0)** — Porta **Pauli-X** (NOT quântica) aplicada ao qubit `q0`.  
   Essa operação inverte o estado de $|0\rangle$ para $|1\rangle$:
   $$
   X|0\rangle = |1\rangle, \quad X|1\rangle = |0\rangle
   $$
   Após essa porta, o sistema está em \(|10⟩\).

2. **H(0)** — Porta **Hadamard** aplicada a `q0`.  
   Essa porta cria uma superposição linear entre os estados base do qubit:
   $$
   H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}
   $$
   Assim, o estado do sistema torna-se:
   $$
   |\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |10\rangle)
   $$

3. **CX(0, 1)** — Porta **CNOT**, com `q0` como qubit de controle e `q1` como alvo.  
   Essa operação inverte o segundo qubit (`q1`) apenas quando o controle (`q0`) está em \(|1⟩\):
   $$
   |00\rangle \rightarrow |00\rangle, \quad |10\rangle \rightarrow |11\rangle
   $$
   Aplicando isso ao estado anterior, obtemos:
   $$
   |\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle)
   $$
   Esse é um dos **estados de Bell**, ou **estado entrelaçado**.

4. **R_y(π)** — Rotação de \(\pi\) radianos em torno do eixo **y** aplicada ao qubit `q1`.  
   A matriz dessa operação é:
   $$
   R_y(\pi) =
   \begin{bmatrix}
   0 & -1 \\
   1 & 0
   \end{bmatrix}
   $$
   Essa rotação troca os estados $|0⟩$ e $|1⟩$, invertendo o segundo qubit:
   $$
   |\psi\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle)
   $$
   Esse é outro estado de Bell (o estado $|\Psi^-\rangle$).

5. **H(0)** — Aplicação final da porta **Hadamard** ao qubit `q0`.  
   Essa etapa transforma novamente a base de superposição do primeiro qubit, alterando a estrutura de interferência entre os componentes do estado. Dependendo das fases relativas e do efeito das rotações anteriores, o resultado final mantém coerência, mas modifica as amplitudes do sistema para o próximo passo de medição.

---

### Estado final esperado

Após toda a sequência de operações, o sistema encontra-se em um **estado entrelaçado com fase relativa** entre $|01\rangle$ e $|10\rangle$:

$$
|\psi_{\text{final}}\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle)
$$

Esse tipo de estado é essencial em protocolos de **teletransporte quântico**, **distribuição de chaves quânticas (QKD)** e **porta lógica controlada universal**, pois demonstra a correlação não clássica entre os qubits.

---

### Interpretação do circuito

- O qubit **q0** é usado como **controle lógico** de operações subsequentes.  
- O qubit **q1** atua como **alvo** das rotações e da operação CNOT.  
- As duas portas Hadamard em `q0` criam e depois reconfiguram a base de superposição.  
- A porta **Ry(π)** é responsável por introduzir uma rotação contínua (analógica) na esfera de Bloch, produzindo interferência de fase.

Em conjunto, essas operações geram um circuito capaz de **preparar, manipular e medir estados de Bell** — um dos blocos fundamentais da computação e comunicação quântica.
